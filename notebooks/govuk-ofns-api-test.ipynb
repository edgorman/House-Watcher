```python
wards = []
wards.clear()
wards.append(ofns_api.get_wards_from_district('Cherwell'))
print(wards[0])
print(len(wards[0]))
```

```python
coordinates= []
coordinates.clear()

for j in range (0,len(wards[0])):
    coordinates.append(ofns_api.get_ward_polygon('Cherwell', wards[0][j]))

#print(coordinates[0])
#print("\n")
#print(coordinates[1])
#print("\n")
print(coordinates)
```

```python
wards_neighbour = []
wards_neighbour.clear()
wards_neighbour_count = []
wards_neighbour_count.clear()
average_price = []
average_price.clear()

for n in range (0,4):
    wards_neighbour.clear()
    wards_neighbour_count.clear()
    average_price.clear()
    for b in range (0,len(coordinates)):
        wards_neighbour.append([])
        c=0
        for a in range (0,len(coordinates)):
            wards_neighbour[b].insert(a,0)
            for j in range (0,len(coordinates[a])):
                for k in range (0,len(coordinates[b])):
                    if a == b:
                        break
                    elif coordinates[a][j] == coordinates[b][k]:
                        wards_neighbour[b].insert(a,200) #price_mean[n][b]
                        if wards_neighbour[b][a] != 0:
                            c+=1
                        break
                else:
                    continue
                break
        wards_neighbour_count.insert(b,c)
        average_price.insert(b,int(float((sum(wards_neighbour[b]))/wards_neighbour_count[b])))

    #for j in range(0,len(coordinates)):
        #if (price_mean[n][j] == 0):
            #price_mean[n].insert(j,average_price[j])
        
print(wards_neighbour)
print(wards_neighbour_count)
print(average_price)
print(len(wards_neighbour))
print(len(wards_neighbour_count))
print(len(average_price))
```
